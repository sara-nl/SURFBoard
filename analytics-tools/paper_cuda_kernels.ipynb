```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet('data/pcm/pyprof_kernels/run_0_config_8_2_ib_nvprof_2_resnet50_dali-cpu-to-gpu_32_fp16_fp32_ret_0_0.gzip')
```

```python
def kernels(df):
    df_forward = df[['Kernel', 'Sil(ns)']].groupby('Kernel').sum()
    return df_forward.sort_values(by='Sil(ns)', ascending=False)
```

```python
top_kernels = kernels(df)*1E-9/25
top_kernels['%'] = 100*top_kernels['Sil(ns)']/top_kernels['Sil(ns)'].sum()
top_kernels.columns = ['duration (s)', '%']
```
